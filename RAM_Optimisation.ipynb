```python
import os
import sys
import pandas as pd
import joblib
import torch
import importlib.util
# Define paths based on our setup (Data & Plots saved to Drive)
DRIVE_DIR = '/content/drive/MyDrive/ram_optimisation'
DATA_DIR = os.path.join(DRIVE_DIR, 'data')
PLOTS_DIR = os.path.join(DRIVE_DIR, 'plots')
LOCAL_DIR = os.getcwd() # Where your scripts (config.py, etc.) live
print("="*70)
print(" PROJECT PIPELINE COMPLETED SUCCESSFULLY!")
print("="*70)
print(f" Data Directory:  {DATA_DIR}")
print(f" Plots Directory: {PLOTS_DIR}\n")
# --- 1. File Summary ---
if os.path.exists(DATA_DIR):
    data_files = [f for f in os.listdir(DATA_DIR) if not f.startswith('.')]
    print(f" Data Files Generated ({len(data_files)}):")
    for f in sorted(data_files):
        # Highlight classification/DNN files
        if "classification" in f or "dnn" in f:
            print(f"    {f}")
        # Mark the obsolete regression files
        elif "regression" in f:
            print(f"    {f} (Obsolete task)")
        else:
            print(f"    {f}")
if os.path.exists(PLOTS_DIR):
    plot_files = [f for f in os.listdir(PLOTS_DIR) if not f.startswith('.')]
    print(f"\n Visualizations Generated ({len(plot_files)}):")
    for f in sorted(plot_files):
        print(f"    {f}")
print("\n" + "="*70)
print(" FINAL CLASSIFICATION SUMMARY")
print("="*70)
# --- 2. Load and Display Metrics ---
# Load DNN Results
try:
    dnn_res = joblib.load(os.path.join(DATA_DIR, 'dnn_results.pkl'))
    if 'classification_threshold' in dnn_res:
        thresh = dnn_res['classification_threshold']
        print(f" High Absorber Threshold (Median e_total): > {thresh:.4f}\n")
    if 'final_metrics' in dnn_res:
        acc = dnn_res['final_metrics'].get('Accuracy', 0)
        f1 = dnn_res['final_metrics'].get('F1', 0)
        print(f" Deep Neural Network (RAMNet):")
        print(f"   • Accuracy: {acc*100:.1f}%")
        print(f"   • F1 Score: {f1:.4f}")
except FileNotFoundError:
    print(" DNN results not found. Did you run Step 6?")
# Load Classical Classification Results
try:
    class_res = joblib.load(os.path.join(DATA_DIR, 'classification_results.pkl'))
    print(f"\n🤖 Classical Models (Accuracy):")
    for model_name, metrics in class_res.items():
        acc = metrics.get('accuracy', 0)
        print(f"   • {model_name}: {acc*100:.1f}%")
except FileNotFoundError:
    print("\n Classical classification results not found. Did you run Step 5?")
# --- 3. Top 5 Material Candidates ---
print("\n" + "="*70)
print(" TOP 5 RADAR ABSORBING MATERIAL CANDIDATES")
print("="*70)
try:
    # Append local dir to path to import config
    if LOCAL_DIR not in sys.path:
        sys.path.append(LOCAL_DIR)
    from config import FEATURE_COLS, TARGET_COL, HIDDEN_DIMS, DROPOUT, DEVICE
    # Robust import for RAMNet
    dl_path = os.path.join(LOCAL_DIR, '6_deep_learning.py')
    spec = importlib.util.spec_from_file_location("deep_learning", dl_path)
    dl_module = importlib.util.module_from_spec(spec)
    spec.loader.exec_module(dl_module)
    RAMNet = dl_module.RAMNet
    features_path = os.path.join(DATA_DIR, 'features.csv')
    model_path = os.path.join(DATA_DIR, 'dnn_model.pth')
    if os.path.exists(features_path) and os.path.exists(model_path):
        df = pd.read_csv(features_path)
        # Initialize model and load weights
        input_dim = len(FEATURE_COLS)
        model = RAMNet(input_dim, HIDDEN_DIMS, DROPOUT).to(DEVICE)
        model.load_state_dict(torch.load(model_path, map_location=DEVICE))
        model.eval()
        # Run predictions
        X_tensor = torch.tensor(df[FEATURE_COLS].values, dtype=torch.float32).to(DEVICE)
        with torch.no_grad():
            logits = model(X_tensor)
            probs = torch.sigmoid(logits).cpu().numpy().flatten()
        df['Probability'] = probs
        top_5 = df.sort_values('Probability', ascending=False).head(5)
        print(f"{'Rank':<6} {'Formula':<20} {'Probability':<15} {'Actual e_total':<15}")
        print("-" * 70)
        for i, (_, row) in enumerate(top_5.iterrows(), 1):
            formula = row['formula_pretty']
            prob = row['Probability'] * 100
            actual_e = row[TARGET_COL]
            print(f"{i:<6} {formula:<20} {prob:>10.2f}%     {actual_e:>10.4f}")
    else:
        print(" Could not find features.csv or dnn_model.pth to generate predictions.")
except Exception as e:
    print(f" Could not generate top 5 predictions in this cell. Error: {e}")
print("\n" + "="*70)
print(" Execution finished! Check the plots/ folder for visual insights.")
```
